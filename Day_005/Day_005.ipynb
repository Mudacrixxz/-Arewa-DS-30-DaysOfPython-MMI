```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
ages.sort()
print(ages)
min_value = min(ages)
print('The Min Value is: ', min_value)

max_value = max(ages)
print('The Max Value is', max_value)

sum_ages = sum(ages)  #sum
print('The sum of ages', sum_ages)

average = sum_ages / 10 # Average
print('The Average of ages = ', average)

range_of_ages = max_value - min_value #range
print('The Range of ages = ',range_of_ages)
```

Question (1) cont.

```python
compare1 = min_value - average 
print(abs(compare1))
compare2 = max_value - average
print(abs(compare2))

len_of_ages = len(ages)  #finding the median age
print(len_of_ages)

if len_of_ages % 2 == 0:
	condition1 = ages[len_of_ages // 2]
	condition2 = ages[len_of_ages // 2 - 1]

	median_age = (condition1 + condition2) / 2

else:
	median_age = ages[len_of_ages // 2]

print('The median age = ', median_age)
```
